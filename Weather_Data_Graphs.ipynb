```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load dataset
df = pd.read_csv('/content/DailyDelhiClimateTrain.csv')

# convert date
df['date'] = pd.to_datetime(df['date'])

df.head()
```

1. Temperature Trend (Time Series)

```python
plt.figure()
plt.plot(df['date'], df['meantemp'])
plt.title("Temperature Trend")
plt.xlabel("Date")
plt.ylabel("Mean Temperature")
plt.xticks(rotation=45)
plt.show()
```

2. Humidity Distribution

```python
plt.figure()
sns.histplot(df['humidity'], bins=30)
plt.title("Humidity Distribution")
plt.show()
```

3. Correlation Heatmap

```python
plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

4. Wind Speed vs Temperature

```python
plt.figure()
plt.scatter(df['wind_speed'], df['meantemp'])
plt.xlabel("Wind Speed")
plt.ylabel("Temperature")
plt.title("Wind Speed vs Temperature")
plt.show()
```

5. Rolling Average

```python
df['temp_rolling'] = df['meantemp'].rolling(7).mean()

plt.figure()
plt.plot(df['date'], df['meantemp'], label='Original')
plt.plot(df['date'], df['temp_rolling'], label='7-day Avg')
plt.legend()
plt.title("Temperature Trend with Rolling Average")
plt.show()
```
