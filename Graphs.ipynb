```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/netflix.csv.xlsx")
```

```python
print(df.head())
print(df.info())
```

Movies vs TV Shows

```python
plt.figure()
df["type"].value_counts().plot(kind='bar')
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

Content Growth Over Years

```python
plt.figure()
df["release_year"].value_counts().sort_index().plot()
plt.title("Content Growth Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

Top 10 Countries

```python
plt.figure()
df["country"].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Content Producing Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Ratings Distribution

```python
plt.figure()
df["rating"].value_counts().head(10).plot(kind='bar')
plt.title("Ratings Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Movies vs TV Shows Trend

```python
plt.figure()

movies = df[df["type"] == "Movie"]["release_year"].value_counts().sort_index()
tv = df[df["type"] == "TV Show"]["release_year"].value_counts().sort_index()

plt.plot(movies, label="Movies")
plt.plot(tv, label="TV Shows")

plt.title("Movies vs TV Shows Growth Over Time")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
plt.show()
```

Duration Analysis

```python

df_movies = df[df["type"] == "Movie"].copy()
df_movies["duration"] = df_movies["duration"].str.replace(" min", "").astype(float)

plt.figure()
plt.hist(df_movies["duration"], bins=20)
plt.title("Movie Duration Distribution")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.show()
```
